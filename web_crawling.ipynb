```python
poems = []
response = requests.get("http://tukay.informatikaexpert.ru/SomePoems.html")
soup = BeautifulSoup(response.content, "html.parser")
for i in range(5):
    tag = soup.find("div", id=f"tabs-{i}")
    if tag: 
        title_tag = tag.find("h1")  # Assuming the title is in an <h1> tag within the div
        title = title_tag.get_text() if title_tag else f"Poem {i}"
        content = tag.get_text()
        poems.append({
            "Title": title,
            "Poem": content.strip()  # Strip leading/trailing whitespace
        })
```

```python
all_ahref = []

# Loop through the pages to collect poem URLs
for i in range(9):
    response2 = requests.get(f"http://gabdullatukay.ru/eng/work/page/{i}/")
    soup = BeautifulSoup(response2.content, "html.parser")
    poems = soup.find_all('a', rel='bookmark')
    titles = soup.find_all('h1', class_="entry-title")
    for poem in poems:
        all_ahref.append(poem)

# List to store all poems with their titles
all_poems = []

# Loop through each poem URL to extract title and content
for url in all_ahref:
    response = requests.get(url["href"])
    soup = BeautifulSoup(response.content, 'html.parser')
    title_tag = soup.find('h1', class_='entry-title')
    content_tags = soup.find('p')
    
    if title_tag and content_tags:
        title = title_tag.get_text().strip()
        content = '\n'.join([tag.get_text().strip() for tag in content_tags])
        all_poems.append({
            "Title": title,
            "Poem": content
        })

# Define the filename for the CSV file
filename = 'poems6.csv'

# Specify the fieldnames (headers)
fieldnames = ['Title', 'Poem']


# Write the data to the CSV file
with open(filename, mode='a', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames, quoting=csv.QUOTE_ALL)
    
    for poem in all_poems:
        writer.writerow(poem)
```
